```python
# ランダムな整数を利用するために、randamモジュールをインポート
import random

# 変数varに0～20までのランダムな整数を代入する
var = random.randint(0,20)

# 変数varの値を出力
print(var)

# 変数varの値が3の倍数であれば「Fizz」、5の倍数であれば「Buzz」、3の倍数でありかつ5の倍数であれば「FizzBuzz」、いずれも該当しなければvarの値を出力
if var % 3 == 0 and var % 5 == 0 and var != 0:
    print("「FizzBuzz」")
elif var % 3 == 0 and var != 0:
    print("「Fizz」")
elif var % 5 == 0 and var != 0:
    print("「Buzz」")
else:
    print(var)
```
