```python
# 第一引数に商品の金額、第二引数に消費税を設定する
def tax_included(product_price: int, product_tax: int):

# 与えられた引数product_priceに消費税を加算し、変数totalに代入する
    total = product_price + product_price * product_tax / 100
    return total

# 関数を呼び出して、商品の金額と消費税（10%）を設定する
result = tax_included(110, 10)
print(f"{result}円")
```
