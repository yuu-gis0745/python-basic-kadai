```python
# Humanクラス
class Human:

    # コンストラクタを定義付け
    def __init__(self, name, age):
        # 属性を定義
        self.name = name
        self.age = age

    # printinfoを用いた名前(name)と年齢(age)の標準出力
    def printinfo(self):
        print(self.name)
        print(self.age)

# Humanをインスタンス化
p = Human("侍太郎", 38)

# 「インスタンス化した変数.メソッド」
a = p.printinfo()
```
