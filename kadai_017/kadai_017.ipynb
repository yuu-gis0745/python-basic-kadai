```python
# 空のユーザーリストを作成
user_list = []

class Human:
    # コンストラクタを定義
    def __init__(self,name,age):
        self.name = name
        self.age = age

    # リストに名前と年齢を追加
    def add_to_user_list(self):
        user_list.append(self)

    # check_adultメソッドを定義
    def check_adult(self):
        # 大人かどうかの判断
        if self.age >= 20:
            print(f"{self.name}は大人です")
        else:
            print(f"{self.name}は大人ではありません")

# 複数のインスタンスを作成
o = Human("侍太郎",15)
o.add_to_user_list()
o = Human("侍二郎", 3)
o.add_to_user_list()
o = Human("侍三郎", 23)
o.add_to_user_list()
o = Human("侍花子", 11)
o.add_to_user_list()
o = Human("侍正美", 33)
o.add_to_user_list()
o = Human("侍未希", 18)
o.add_to_user_list()

# check_adultで大人かどうか判定する
for age in user_list:
    age.check_adult()
```
