```python
# .appendをcheck_adultで追加したver2.0
user_names = []
user_ages = []

class Human:
    # コンストラクタを定義
    def __init__(self,name,age):
        self.name = name
        self.age = age

    # check_adultメソッドを定義
    def check_adult(self):
        # リストに名前と年齢を追加
        for i in range(1):
            user_names.append(self.name)
            user_ages.append(self.age)
        # 大人かどうかの判断
        if self.age >= 20:
            print("大人です")
        else:
            print("大人ではありません")

# 複数のインスタンスを作成
o = Human("侍太郎",15)
human = o.check_adult()
o = Human("侍二郎", 3)
human = o.check_adult()
o = Human("侍三郎", 23)
human = o.check_adult()
o = Human("侍花子", 11)
human = o.check_adult()
o = Human("侍正美", 33)
human = o.check_adult()
o = Human("侍未希", 18)
human = o.check_adult()

# リストに代入された要素を確認する
print(user_names)
print(user_ages)
```
